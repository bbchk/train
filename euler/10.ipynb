We need to find all the primes to the certain number, for this we can apply sieve of Eratosthenes, push every prime into array, and then reduce it

Algorithm



```javascript
const limit = 2_000_000

//creating sieve
const sieve = Array.from({ length: limit }, (_, i) => true);

for (let i = 2; i * i <= limit / 2; i++ ) {
    if(sieve[i]) {
        for(let j = i * i; j <= limit; j=+i){
            sieve[j] = false
        }
    }
}


const primes = [];
for (let i = 2; i <= limit; i++) {
    if(sieve[i]){
        primes.push(i);
    }
}
console.log(primes)
```
