```javascript
//lets just try brute force, it will be O(n*log n)

const lim = 1_000_000
let longestChainLen = 0
let longestChainNum = 1

for(let i = 1; i <= lim; i++) {
    let num = i
    let chainLen = 0
    while(num != 1) {
        chainLen++
        if(num % 2 == 0){
            num /= 2
        }else {
            num = num * 3 + 1
        }
    }

    if(chainLen > longestChainLen){
        longestChainLen = chainLen
        longestChainNum = i
    }
}

console.log(longestChainNum)
console.log(longestChainLen)
```

```javascript
//Brute-force with memoization (using object to cache results)
const lim = 1_000_000;
let longestChainLen = 0;
let longestChainNum = 1;
const cache = {};

function collatzLength(num) {
    if (num === 1) return 1; // Base case
    if (cache[num]) return cache[num]; // Return cached result if available

    let nextNum = num % 2 === 0 ? num / 2 : num * 3 + 1;
    cache[num] = 1 + collatzLength(nextNum); // Store the result in cache

    return cache[num];
}

for (let i = 1; i <= lim; i++) {
    let chainLen = collatzLength(i);

    if (chainLen > longestChainLen) {
        longestChainLen = chainLen;
        longestChainNum = i;
    }
}

console.log(longestChainNum);
console.log(longestChainLen);
```
