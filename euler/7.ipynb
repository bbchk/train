```javascript
// cool algorithm to find out if number is prime with complexity O (sqrt(n))

function isPrime(n) {
    if (n <= 1) return false;
    if (n <= 3) return true;
    if (n % 2 == 0 || n % 3 == 0)
        return false;

    for (let i = 5; i * i <= n; i += 6){
        if (n % i == 0 || n % (i + 2)) 
            return false;
    }

    return true;
}
```

```javascript
function sieveOfEratosthenes(limit) {
    // Initialize the array of boolean values
    let sieve = new Array(limit + 1).fill(true);
    sieve[0] = sieve[1] = false; // 0 and 1 are not primes

    // Implement the Sieve of Eratosthenes
    for (let i = 2; i * i <= limit; i++) {
        if (sieve[i]) {
            for (let j = i * i; j <= limit; j += i) {
                sieve[j] = false;
            }
        }
    }

    // Extract the primes
    let primes = [];
    for (let i = 2; i <= limit; i++) {
        if (sieve[i]) {
            primes.push(i);
        }
    }
    
    return primes;
}

// Estimate the limit (this could be adjusted if necessary)
const estimatedLimit = 120000; // An estimated upper bound for the 10,001st prime
const primes = sieveOfEratosthenes(estimatedLimit);

// Ensure we have enough primes and get the 10,001st prime
if (primes.length >= 10001) {
    console.log(primes[10000]); // 10,001st prime (0-based index)
} else {
    console.log("Increase the estimated limit.");
}
```
